### Q4.

Crie uma lista de listas (uma matriz de 2 dimensões) de tamanho $n \times n$, com $n$ dado pelo usuário. Cada elemento da matriz deve ser o produto dos índices da linha pela coluna.

Exemplo:
```
Digite n: 4
Matriz:
[[0, 0, 0, 0],
 [0, 1, 2, 3],
 [0, 2, 4, 6],
 [0, 3, 6, 9]]
```

````python
# Implemente sua solução aqui

# Função para criar a matriz
def cria_matriz(n):
    return [[i * j for j in range(n)] for i in range(n)]

# Pedir o tamanho da matriz ao usuário
n = int(input("Digite o tamanho da matriz (n): "))

# Criar a matriz
matriz = cria_matriz(n)

# Imprimir a matriz
for linha in matriz:
    print(linha)
````

### Q5.

Escreva um programa para encontrar a diferença entre duas listas, incluindo elementos duplicados.

Ex:
```
Listas originais: 
[1, 1, 2, 3, 3, 4, 4, 5, 6, 7]
[1, 1, 2, 4, 5, 6]
Diferença entre as listas:
[3, 3, 4, 7]
```

````python
# Implemente sua solução aqui
def diferenca_listas(lista1, lista2):
    # Cria uma cópia da lista1 para evitar modificar a lista original
    lista1_copia = lista1.copy()

    # Remove os elementos da lista2 da lista1_copia
    for elemento in lista2:
        if elemento in lista1_copia:
            lista1_copia.remove(elemento)

    return lista1_copia

# Listas originais
lista1 = [1, 1, 2, 3, 3, 4, 4, 5, 6, 7]
lista2 = [1, 1, 2, 4, 5, 6]

# Encontra a diferença entre as listas
diferenca = diferenca_listas(lista1, lista2)

print("Listas originais:")
print("Lista 1:", lista1)
print("Lista 2:", lista2)
print("Diferença entre as listas:", diferenca)

````

```
Tamanho para divisão: 4

Tamanho para divisão: 0
```
